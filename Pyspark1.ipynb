```python
import pyspark
from pyspark.sql import SparkSession
from pyspark import SparkContext, SparkConf
spark = SparkSession.builder.appName('RDD example').getOrCreate()
pyspark.sql.session.SparkSession
```

```python
df2 = spark.read.text("bank.txt")
```

```python
df2.collect()
```

```python
bank = spark.sparkContext.textFile('bank.txt')
```

```python
bank.collect()
```

```python
clean = bank.map( lambda d :(d.split('\t')))
clean.collect()
```

```python
clean.collect()[0][1]
```

```python
clean.map(lambda d:d[1]).distinct().count()
```

how many distinct branches are there and how many distinct customers are there for each branch

number of credit transactions

```python
clean.filter(lambda d:d[3]=='C').count()
```

how many credit transactions per customer

```python
clean.filter(lambda d:(d[1]=='123876122')& (d[3]=='C')).count()
```

```python
oneMoreVariable = clean.map(lambda d:(d[1], d[2].split('-')[2], d[3]))
oneMoreVariable.collect()
```
