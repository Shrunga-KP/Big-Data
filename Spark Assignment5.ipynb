```python
import pyspark
from pyspark.sql import SparkSession
from pyspark import SparkContext, SparkConf
spark = SparkSession.builder.appName('RoadAccident DF').getOrCreate()
pyspark.sql.session.SparkSession
```

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
data = spark.read.csv("drive/My Drive/Colab Notebooks/Data/Road_accidents.csv", header=True,inferSchema=True)
data.show()
data.printSchema()
```

```python
data_time = spark.read.csv("drive/My Drive/Colab Notebooks/Data/Road_accidents_time.csv", header=True,inferSchema=True)
data_time.show()
data_time.printSchema()
```

```python
#1. Add column 'Total', giving sum of accidents in each month
total = data.withColumn('TOTAL',(col('JANUARY')+col('FEBRUARY')+col('MARCH')+col('APRIL')+col('MAY')\
    +col('JUNE')+col('JULY')+col('AUGUST')+col('SEPTEMBER')+col('OCTOBER')+col('NOVEMBER')+col('DECEMBER')))
total.show()
```

```python
#2. Which state has highest number of accidents in year 2013?
highest_acc = total.select('STATE/UT','YEAR','TOTAL').filter(col('YEAR') == 2013)
highest_acc.where(col('TOTAL') == highest_acc.agg({'TOTAL':'max'}).collect()[0][0]).show()
```

```python
#3. Find the average monthly accidents for each state.
monthAcc = data.withColumn('AVG',(col('JANUARY')+col('FEBRUARY')+col('MARCH')+col('APRIL')+col('MAY')+col('JUNE')+col('JULY')+col('AUGUST')
    +col('SEPTEMBER')+col('OCTOBER')+col('NOVEMBER')+col('DECEMBER'))/12)
monthAcc.select('STATE/UT','AVG').groupBy('STATE/UT','AVG').count().distinct().show(30)
```

```python
#4. Which month has highest accidents?

highest_acc = data.select([col for col in data.columns if col!= 'STATE/UT' and col!='YEAR'])
var  = {}
for i in highest_acc.columns:
  var[i] = highest_acc.agg({i:'max'}).collect()[0][0]
data2 = spark.createDataFrame(var.items(), ["month", "value"])
maxValue = data2.agg({'value':'max'}).collect()[0][0]
data2.select('month','value').where(col('value')==maxValue).show()
```

```python
#6. Which state has more accidents from year 2001 to 2014?

more_acc = total.select('TOTAL','YEAR','STATE/UT').where(col('YEAR') <= 2014)
more_accident = more_acc.groupBy('STATE/UT','TOTAL').count().agg({'TOTAL':'sum'}).collect()[0][0]
more_accident
```

```python
#8. Which month has more accidents over the year?
highest_acc = data.select([col for col in data.columns if col!= 'STATE/UT' and col!='YEAR'])
var  = {}
for i in highest_acc.columns:
  var[i] = highest_acc.agg({i:'max'}).collect()[0][0]
df4 = spark.createDataFrame(var.items(), ["month", "value"])
maxValue = df4.agg({'value':'max'}).collect()[0][0]
df4.select('month','value').where(col('value')==maxValue).show()
```
